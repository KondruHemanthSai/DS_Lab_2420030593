```python
import pandas as pd
from scipy.stats import chi2_contingency
data = pd.DataFrame({
    'Gender': ['Male','Female','Male','Female','Male','Female'],
    'Purchase': ['Yes','No','Yes','No','No','Yes']
})
table = pd.crosstab(data['Gender'], data['Purchase'])
print("Contingency Table:")
print(table)
chi2, p, dof, expected = chi2_contingency(table)
print("\nChi-Square Statistic:", chi2)
print("P-value:", p)
if p < 0.05:
    print("Significant Association Exists")
else:
    print("No Significant Association")
```

```python
from scipy.stats import ttest_ind
group1 = [85, 88, 90, 92, 87]
group2 = [78, 80, 82, 79, 81]
t_stat, p_value = ttest_ind(group1, group2)
print("T-Statistic:", t_stat)
print("P-value:", p_value)
if p_value < 0.05:
    print("Means are significantly different")
else:
    print("No significant difference")
```

```python
from scipy.stats import f_oneway
group1 = [85,88,90,92,87]
group2 = [78,80,82,79,81]
group3 = [65,68,70,72,69]
f_stat, p_value = f_oneway(group1, group2, group3)
print("F-Statistic:", f_stat)
print("P-value:", p_value)
if p_value < 0.05:
    print("At least one group mean differs")
else:
    print("No significant difference")
```

```python
import numpy as np
from scipy.stats import norm
sample = np.array([70,72,68,75,74,71,69,73,70,72])
mu0 = 70
sample_mean = sample.mean()
sample_std = sample.std(ddof=1)
z_stat = (sample_mean - mu0) / (sample_std / np.sqrt(len(sample)))
p_value = 2 * (1 - norm.cdf(abs(z_stat)))
print("Z-Statistic:", z_stat)
print("P-value:", p_value)
if p_value < 0.05:
    print("Sample mean differs from population mean")
else:
    print("No significant difference")
```

```python
import pandas as pd
import numpy as np
from scipy.stats import zscore
data = pd.DataFrame({
    'Marks':[45,50,52,48,49,51,47,46,150]
})
data['Z_Score'] = zscore(data['Marks'])
outliers = data[np.abs(data['Z_Score']) > 3]
print("Dataset:")
print(data)
print("\nOutliers:")
print(outliers)
```

```python
import pandas as pd
data = pd.DataFrame({
    'Marks':[45,50,52,48,49,51,47,46,150]
})
Q1 = data['Marks'].quantile(0.25)
Q3 = data['Marks'].quantile(0.75)
IQR = Q3 - Q1
lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR
outliers = data[
    (data['Marks'] < lower) |
    (data['Marks'] > upper)
]
print("Q1:", Q1)
print("Q3:", Q3)
print("IQR:", IQR)
print("\nOutliers:")
print(outliers)
```
